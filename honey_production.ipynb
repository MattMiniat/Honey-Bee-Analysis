```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
file_to_load = "Honey_Production.csv"
file2_to_load = "Honey_Prices.csv"
production_data = pd.read_csv(file_to_load)
price_data = pd.read_csv(file2_to_load)
```

```python
honey_df = pd.DataFrame(production_data)
```

```python
honey_df["Value"] = pd.to_numeric(honey_df["Value"].str.replace(',',''), errors='coerce')
production = honey_df[["Year" , "Value"]]
production
```

```python
production_plot = production.plot.line(x='Year', y='Value', xticks= production['Year'])
production_plot.set_title("U.S. Honey Production 1997 - 2017 (USDA Census)")
production_plot.set_xlabel("Year")
production_plot.set_ylabel("Production (per 100,000,000 lbs)")
```

```python
Honey_prices = pd.DataFrame(price_data)
```

```python
National_prices = Honey_prices.loc[(Honey_prices["State"] == "US TOTAL")]
```

```python
Prices = National_prices[["Year", "Value"]]
```

```python
Real_Prices = round(Prices["Value"]/100,2)
Prices["Value"] = Real_Prices

```

```python
# price_plot = Prices.plot.line(x='Year', y='Value', xticks= production['Year'])
price_plot = Prices.plot.line(x='Year', y='Value')
price_plot.set_title("U.S. Honey Prices 1987 - 2018 (USDA Survey)")
price_plot.set_xlabel("Year")
price_plot.set_ylabel("Price per $1 USD")
```
